### Find out optimum number of clusters, displaying elbow plot for k value

####
- You will notice that the inertia is the sum of squared distance between each point and its nearest cluster center - Euclidiean distance

```python
X1 = df.loc[:, ['Age', 'Spending Score (1-100)']].values
from sklearn.cluster import KMeans
wcss = []
# wcss - within cluster sum of square
for k  in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
    #print(wcss)
plt.figure(figsize=(12, 6))
plt.grid()
plt.plot(range(1, 11), wcss, linewidth = 2, color='red', marker=8)
plt.xlabel("K value ")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
labels = kmeans.fit_predict(X1)
print(labels)

#Now each cluster is labelled with its approriate value 
```

```python
print(kmeans.cluster_centers_) # centroid - a center value for each cluster of 4
# 27.61 is for Age value and 49.14 is Spending score value
```

```python
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black')
plt.title("Cluster of customers")
plt.xlabel("Age")
plt.ylabel("Spending score")
plt.show()
```
